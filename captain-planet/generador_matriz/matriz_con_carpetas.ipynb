```python
import pandas as pd

matriz = pd.read_csv(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\data\clean\matriz_4.csv', encoding='utf-8')
carpetas = pd.read_csv(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\data\raw\df_victimas.csv', encoding='utf-8')
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python

carpetas[["fechainicio"]] = carpetas[["fechainicio"]].apply(pd.to_datetime, errors = 'coerce', infer_datetime_format=True)

carpetas["fechainicio"] = carpetas["fechainicio"].dt.strftime('%Y-%m')

carpetas_agrupadas = carpetas.groupby(['alcaldiahechos','fechainicio','delito']).count()

carpetas_agrupadas.reset_index(inplace=True)
```

```python
carpetas_agrupadas = carpetas_agrupadas[['alcaldiahechos','fechainicio','delito','idcarpeta']]
carpetas_agrupadas.rename(columns = {'idcarpeta':'num_carpetas'}, inplace = True)

carpetas_agrupadas
```

```python
prueba_merge = matriz.merge(carpetas_agrupadas, left_on=['alcaldia', 'anio-mes-admision/radicacion'], right_on=['alcaldiahechos','fechainicio'])
```

```python
prueba_merge.to_csv(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\data\clean\matriz_con_carpetas.csv')
```
