```python
#base de datos con densidad poblacional por colonia
from scipy.fft import idst
from simpledbf import Dbf5
import pandas as pd
import recordlinkage
import funciones
```

```python
# Densidad De Vivienda 2010:
dbf = Dbf5(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\data\raw\ri_11.dbf')
densidad = dbf.to_dataframe().drop_duplicates().dropna()
funciones.estandarizacion(densidad, 'colonia', 'alcaldia')
densidad = densidad[['alcaldia', 'colonia', 'pob_2010', 'SUP_COL_M2', 'DENVIVHa', 'cve_col']]
densidad.rename(columns={'pob_2010':'poblacion_2010', 'SUP_COL_M2':'superficie_colonia_2010', 'DENVIVHa':'densidad_vivienda_2010'}, inplace=True)
densidad['clave'] = densidad['alcaldia'] + " " + densidad['colonia']
del densidad['alcaldia']
del densidad['colonia']
densidad = densidad.sort_values('clave')
```

```python
 # Denuncias con toda la limpieza:

colonias_matched = pd.read_csv(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\analisis_maltrato_animal\data\clean\colonias_matched.csv', encoding='utf-8')
colonias_matched = colonias_matched[['colonia_vieja', 'colonia_nueva', 'alcaldia']]
denuncias = pd.read_csv(r'C:\Users\samue\OneDrive\Documents\GitHub\captain_planet_sammy\captain-planet\data\raw\df_paot_con_limpieza.csv', encoding='utf-8')
funciones.estandarizacion(denuncias, 'colonia', 'alcaldia')

alcaldias = list(colonias_matched.alcaldia.unique())
for alcaldia in alcaldias:
    for index, row in colonias_matched[colonias_matched['alcaldia']==alcaldia].iterrows():
        denuncias.loc[(denuncias['colonia']==row['colonia_vieja']) & (denuncias['alcaldia']==alcaldia),'colonia']=\
            row['colonia_nueva']

denuncias['clave'] = denuncias['alcaldia'] + " " + denuncias['colonia']
del denuncias['alcaldia']
del denuncias['colonia']
```

```python
 # Juntar los datasets:
indexer = recordlinkage.Index()
indexer.block('clave')
candidate_links = indexer.index(denuncias, densidad)

compare = recordlinkage.Compare()

compare.string('clave', 'clave', method='damerau_levenshtein', threshold=0.85)
# compare.string('streetname', 'streetname', method='damerau_levenshtein', threshold=0.7)

# The comparison vectors
compare_vectors = compare.compute(candidate_links, denuncias, densidad)

# compare_vectors.columns
compare_vectors[compare_vectors[0] == 1]
```
