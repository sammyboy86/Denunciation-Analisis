```python
#genera diccionario con meses por anio con columna y valor dados
def totales_vs_concluidas(columna, valor, anio_ini, anio_final):

    denuncias = df_paot[df_paot[columna]==valor]

    denuncias_concluidas_por_mes = {}
    denuncias_totales_por_mes = {}

    for a in range(anio_ini,anio_final):
        temp = {}
        temp2 = {}
        
        for m in range(1,13):
            temp[m] =(denuncias[(denuncias['fecha_de_recepcion'].dt.year==a) & 
            (denuncias['fecha_de_recepcion'].dt.month == m)].shape[0])
            temp2[m] = (denuncias[(denuncias['fecha_de_recepcion'].dt.year==a) & 
            (denuncias['fecha_de_recepcion'].dt.month == m) & (denuncias['estatus']=='Concluida')].shape[0])
        denuncias_totales_por_mes[a] = temp
        denuncias_concluidas_por_mes[a] = temp2
    
    return([denuncias_totales_por_mes, denuncias_concluidas_por_mes, columna, valor, anio_ini, anio_final])

#grafica a partir de dicho diccionario
def graficar(denuncias_por_mes):
    
    totales_por_mes = denuncias_por_mes[0]
    concluidas_por_mes = denuncias_por_mes[1]
    ao_mes = []
    total_concluidas = []
    total_denuncias = []

    for v in totales_por_mes:
        for u in concluidas_por_mes[v]:
            ao_mes.append(str(v) + "-" + str(u))
            total_denuncias.append(totales_por_mes[v][u])
            total_concluidas.append(concluidas_por_mes[v][u])



    dict = {'ao_mes':ao_mes, 'denuncias':total_concluidas, 'tipo':'concluidas'}
    dict2 = {'ao_mes':ao_mes, 'denuncias':total_denuncias, 'tipo':'totales'}



    areas_naturales = pd.DataFrame(dict2)
    areas_naturales2 = pd.DataFrame(dict)

    data_frame= pd.merge(areas_naturales, areas_naturales2, how='outer')

    title = denuncias_por_mes[3] + ' ' + str(denuncias_por_mes[4]) + '-' + str(denuncias_por_mes[5])

    
    fig = px.line(data_frame, x='ao_mes', y='denuncias', color='tipo', title=title)
    fig.show()

    


graficar(totales_vs_concluidas('materia','Maltrato Animal', 2019, 2020))
```
